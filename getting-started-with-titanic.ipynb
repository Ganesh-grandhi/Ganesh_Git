```python
train_data=pd.read_csv("/kaggle/input/titanic/train.csv")
test_data=pd.read_csv("/kaggle/input/titanic/test.csv")
train_data.head()
test_data.head()
```

```python
import seaborn as sns
```

```python
sns.heatmap(train_data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
def fun(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
        
    else:
        return Age
    
```

```python
def fun2(col):
    Q=col[0]
    
    if pd.isnull(Q):
        return 0
    else:
        return Q
```

```python
test_data['Fare']=test_data[['Fare']].apply(fun2,axis=1)
```

```python
test_data.info()
```

```python
train_data['Age']=train_data[['Age','Pclass']].apply(fun,axis=1)
```

```python
test_data["Age"]=test_data[["Age","Pclass"]].apply(fun,axis=1)
```

```python
test_data.info()
```

```python
sex=pd.get_dummies(train_data['Sex'])
embark=pd.get_dummies(train_data['Embarked'])
```

```python
sex_t=pd.get_dummies(test_data['Sex'])
embark_t=pd.get_dummies(test_data['Embarked'])
```

```python
train_data=pd.concat([train_data,sex,embark],axis=1)
```

```python
test_data=pd.concat([test_data,sex_t,embark_t],axis=1)
```

```python
train_data.drop(['Name','Ticket','Sex','Embarked','Cabin','male','S'],axis=1,inplace=True)
```

```python
test_data.drop(['Name','Ticket','Sex','Embarked','Cabin','male','S'],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
y = train_data["Survived"]
X=train_data.drop('Survived',axis=1)

#features = ["Pclass", "Sex", "SibSp", "Parch"]
#X = pd.get_dummies(train_data[features])
#X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=1000, random_state=20)
#model=LogisticRegression()
model.fit(X, y)
predictions = model.predict(test_data)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('my_submission.csv', index=False)
print("Your submission was successfully saved!")
```
